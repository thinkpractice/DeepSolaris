```python
import cv2
import os
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_path = r"/media/tim/Data/Work/CBS/Test"
paths = [os.path.join(image_path, filename) for filename in os.listdir(image_path) if "mask" not in filename]
```

```python
images = [cv2.imread(path) for path in paths]
```

```python
_, ax = plt.subplots(5, 5, figsize=(16,16))
i = 25
for r in range(5):
    for c in range(5):
        ax[r,c].imshow(images[i][:,:, ::-1])
        i += 1
```
