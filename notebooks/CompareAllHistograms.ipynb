```python
from ProjectPaths import ProjectPaths
from DirectoryFilter import DirectoryFilter
from Histograms import Histograms
import os
import cv2 as cv
import numpy as np
```

```python
directoryFilter = DirectoryFilter(os.path.join(ProjectPaths.instance().image_dir, "Tiles"))
```

```python
years = ["2013", "2014", "2015", "2016", "2017", "2018"]
```

```python
filenames_per_year = {year : directoryFilter.dir(year).rgb.images.paths for year in years}
filenames_per_year["2013"][:10], filenames_per_year["2018"][:10]
```

```python
def images_for_year(images):
    return [cv.imread(image) for image in images]
```

```python
histogram_per_year = {year : Histograms.summed_hist_for_images(images_for_year(filenames_per_year[year])) for year in filenames_per_year.keys()}
```

```python
histogram_per_year["2013"]
```

```python
def plot_histogram(axes, summed_hist, title):
    axes.set_title(title)
    axes.set_xlim(0, 256)
    axes.set_ylim(0, 400)
    for i, c in enumerate(["b", "g", "r"]):
        axes.plot(summed_hist[i, :], color=c)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        year = years[i]
        plot_histogram(ax[r, c], histogram_per_year[year], year)
        i += 1
```

```python
def one_color_histogram(histogram_per_year, color_index, title):
    for i, color in enumerate(["r", "g", "b", "y", "m", "k"]):
        year = years[i]
        plt.plot(histogram_per_year[year][color_index,:], color=color, label = year)
        plt.title(title)
        plt.legend() 
```

```python
#Draw blue histograms
one_color_histogram(histogram_per_year, 0, "Blue histograms")
```

```python
#Draw green histograms
one_color_histogram(histogram_per_year, 1, "Green histograms")
```

```python
#Draw red histograms
one_color_histogram(histogram_per_year, 2, "Red histograms")
```

```python
first_images = [cv.imread(filenames_per_year[year][1]) for year in years]
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        ax[r,c].imshow(first_images[i][:,:,::-1])
        ax[r,c].set_title(years[i])
        i += 1
```

```python
first_image_hists = [Histograms.all_bands_hist(image) for image in first_images]
```

```python
#Equalize histograms
def equalize_image(image):
    hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)
    equ_image = cv.equalizeHist(hsv_image[:,:,0])
    equ_hsv_image = np.stack([equ_image, hsv_image[:,:,1], hsv_image[:,:,2]], axis=2)
    return cv.cvtColor(equ_hsv_image, cv.COLOR_HSV2BGR)
```

```python
equ_images = [equalize_image(image) for image in first_images]
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        ax[r,c].imshow(equ_images[i][:,:,::-1])
        ax[r,c].set_title(years[i])
        i += 1
```

```python
equ_first_image_hists = [Histograms.all_bands_hist(image) for image in equ_images]
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        ax[r,c].plot(first_image_hists[i], color='b')
        ax[r,c].plot(equ_first_image_hists[i], color='r')
        ax[r,c].set_title(years[i])
        i += 1
```

```python
#CLAHE Equalize histograms
def clahe_equalize_image(image):
    hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)
    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    equ_image = clahe.apply(hsv_image[:,:,0])
    equ_hsv_image = np.stack([equ_image, hsv_image[:,:,1], hsv_image[:,:,2]], axis=2)
    return cv.cvtColor(equ_hsv_image, cv.COLOR_HSV2BGR)
```

```python
clahe_equ_images = [clahe_equalize_image(image) for image in first_images]
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        ax[r,c].imshow(clahe_equ_images[i][:,:,::-1])
        ax[r,c].set_title(years[i])
        i += 1
```

```python
all_equ_histograms = {year : [equalize_image(image) for image in images_for_year(filenames_per_year[year])] for year in years}
```

```python
all_equ_images = all_equ_histograms
all_equ_histograms = { year: Histograms.summed_hist_for_images(all_equ_images[year]) for year in years}
```

```python
_, ax = plt.subplots(2,3, figsize=(10,10))
i = 0
for r in range(2):
    for c in range(3):
        year = years[i]
        plot_histogram(ax[r, c], all_equ_histograms[year], year)
        i += 1
```
