```python
import json
from pprint import pprint
from ProjectPaths import ProjectPaths


with open(ProjectPaths.instance().file_in_base_dir("deepsolaris_annotations.json")) as panelFile:
    panel_annotations = json.load(panelFile)
pprint(panel_annotations)
```

```python
annotations = [annotation for annotation in panel_annotations if len(annotation["annotations"]) > 0]
pprint(annotations)
```

```python
from skimage.draw import polygon
import numpy as np

def get_coords(annotation_coords):
    return [int(float(c)) for c in annotation_coords.split(";")]

def drawPolygon(tileImage, xCoords, yCoords):
    rr, cc = polygon(yCoords, xCoords)
    tileImage[rr, cc] = 255

def drawMaskImage(annotations, tileImageSize=(75,75)):
    image = np.zeros(tileImageSize, dtype=np.uint8)
    for annotation in annotations:
        xCoords = get_coords(annotation["xn"])
        yCoords = get_coords(annotation["yn"])        
        drawPolygon(image, xCoords, yCoords)
    return image
  

image = drawMaskImage(annotations[0]["annotations"])
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.imshow(image)
```

```python
import os

mask_images = []
for annotation in annotations:
    filename, ext = os.path.splitext(annotation["filename"])
    mask_filename = filename + "_mask" + ext
    mask_images.append(drawMaskImage(annotation["annotations"]))
```

```python
np.save(ProjectPaths.instance().file_in_image_dir("ds_image_masks.npy"), mask_images)
```

```python
_, ax = plt.subplots(5,5, figsize=(10,10))

i = 0
for r in range(5):
    for c in range(5):
        ax[r, c].imshow(mask_images[i])
        i += 1        
```
