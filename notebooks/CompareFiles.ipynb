```python
def getProjectionCoords(dataset, x, y):
        """This method converts the pixel coordinate (x, y) into the coordinate system used
        by the projection of the map.
        The equations used are:
        Xp = transform[0] + x * transform[1] + y * tranform[2]
        
        Yp = transform[3] + x * transform[4] + y * transform[5]
        
        :param int x: the pixel x coordinate
        
        :param int y: the pixel y coordinate
        :return: a tuple with the coordinates in projection space.
        :rtype: tuple (double, double)
        """
        transform = dataset.GetGeoTransform()
        Xp = transform[0] + x * transform[1] + y * transform[2]
        Yp = transform[3] + x * transform[4] + y * transform[5]
        return (Xp, Yp)
```

```python
def getBoundingBox(dataset):
    long1, lat1 = getProjectionCoords(dataset, 0, dataset.RasterYSize)
    long2, lat2 = getProjectionCoords(dataset, dataset.RasterXSize, 0)
    return (long1, lat1, long2, lat2)    
```

```python
import numpy as np

def is_bbox_equal(bbox_left, bbox_right):
    return np.allclose(bbox_left, bbox_right)
```

```python
def calc_bboxes(image_paths):
    bboxes = []
    for image_path in image_paths:
        dataset = gdal.Open(image_path)
        bboxes.append((image_path, getBoundingBox(dataset)))
    return bboxes    
        
```

```python
from scipy.spatial import distance

def calc_smallest_distance(bbox, image_bboxes):
    distances = [(image_path, distance.euclidean(bbox, image_bbox)) for image_path, image_bbox in image_bboxes]
    distances.sort(key=lambda x: x[1])
    return distances[0]
```

```python
def calc_smallest_distances(diff_image_paths, compare_image_paths):
    distances = []
    for image_path in diff_image_paths:
        dataset = gdal.Open(image_path)
        bbox = getBoundingBox(dataset)        
        d = calc_smallest_distance(bbox, compare_image_paths)
        distances.append(d)
    return distances
```
