# Check combined histogram

To check whether there are intensity differences between the pictures of Heerlen taken in 2016 and those taken in 2017, we are going to calculate a combined RGB histogram for both of these years. We will then compare then per color value R, G, and B.

```python
def normalized_hist(hist_image, channel, mask=None):
    hist = cv.calcHist([hist_image], [channel], mask, [256], [0, 255])  
    number_of_pixels = hist_image.size if mask is None else mask[mask > 0].size
    return hist / number_of_pixels

def all_bands_hist(hist_image, mask=None):
    all_bands_hist = []
    for i in range(3):
        all_bands_hist.extend(normalized_hist(hist_image, i, mask))
    return np.array(all_bands_hist)
```

```python
hists_2016 = np.array([[normalized_hist(image, channel) for image in images_2016] for channel in range(3)])

hists_2016 = hists_2016.reshape(hists_2016.shape[0:3])
hists_2016.shape
```

```python
summed_hist_2016 = hists_2016.sum(axis=1)
summed_hist_2016.shape
```

```python
for i, c in enumerate(["b", "g", "r"]):
    plt.plot(summed_hist_2016[i, :], color=c)
```

```python
def normalized_hist2(hist_image, channel):
    hist, _ = np.histogram(hist_image[:,:, channel], 256, range=[0, 255])
    number_of_pixels = hist_image.size
    return hist / number_of_pixels
```

```python
hists_2016_2 = np.array([[normalized_hist2(image, channel) for image in images_2016] for channel in range(3)])

hists_2016_2 = hists_2016_2.reshape(hists_2016_2.shape[0:3])
hists_2016_2.shape
```

```python
summed_hist_2016_2 = hists_2016_2.sum(axis=1)
summed_hist_2016_2.shape
```

```python
for i, c in enumerate(["b", "g", "r"]):
    plt.plot(summed_hist_2016_2[i, :], color=c)
```

```python
hists_2017 = np.array([[normalized_hist(image, channel) for image in images_2017] for channel in range(3)])
hists_2017 = hists_2017.reshape(hists_2016.shape[0:3])
hists_2017.shape
```

```python
summed_hist_2017 = hists_2017.sum(axis=1)
for i, c in enumerate(["b", "g", "r"]):
    plt.plot(summed_hist_2017[i], color=c)
```

# Check histogram equalization

From the histograms above, it seems the color intensities between the 2016 and the 2017 pictures of Heerlen varies dramatically. How can we scale both histograms so that color intensities are similar. Would for example histogram equalization work? Because we are working on color images, we need to equalize the intensity channel of the color images, hence we will need to convert a image in BGR color space to a color space that has intensity as one of its channels.

```python
ycrcb = cv.cvtColor(image_2016, cv.COLOR_BGR2YCrCb)
ycrcb.shape
```

```python
def equalize_color_image(image):
    ycrcb = cv.cvtColor(image, cv.COLOR_BGR2YCR_CB)
    channels = cv.split(ycrcb)
    cv.equalizeHist(channels[0],channels[0])
    cv.merge(channels,ycrcb)
    return cv.cvtColor(ycrcb, cv.COLOR_YCR_CB2BGR)
        
```

```python
eq_image_2016 = equalize_color_image(image_2016)
eq_image_2017 = equalize_color_image(image_2017)
eq_image_2016.shape, eq_image_2017.shape
```

```python
_, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(eq_image_2016)
ax[1].imshow(eq_image_2017)
```

```python
eq_hist_2016 = np.array([normalized_hist(eq_image_2016, i) for i in range(3)]).reshape(3, 256)
eq_hist_2017 = np.array([normalized_hist(eq_image_2017, i) for i in range(3)]).reshape(3, 256)
eq_hist_2016.shape, eq_hist_2017.shape
```

```python
_, ax = plt.subplots(3, 1, figsize=(10,10))

for i,c in enumerate(["blue", "green", "red"]):
    ax[i].set_title("{} 2016 vs. {} 2017".format(c,c))
    ax[i].plot(eq_hist_2016[i, :], label="2016")
    ax[i].plot(eq_hist_2017[i, :], color="r", label="2017")
    ax[i].legend()
```

# Mean-center all images

```python
mean = image_2016.mean(axis=(0,1))
```

```python
image_2016 - mean
```

```python
def mean_center(image):
    return image - image.mean(axis=(0,1))

def std(image):
    return image - image.std(axis=(0,1))

def image_center(image):
    return mean_center(image) / std(image)

def image_hist(image, channel):
    hist, _ = np.histogram(image[:,:, channel], 256, [image.min(), image.max()])
    number_of_pixels = image.size
    return hist / number_of_pixels
```

```python
c_2016 = image_center(image_2016)
c_2017 = image_center(image_2017)
```

```python
c_2016.min(), c_2016.max(), c_2017.min(), c_2017.max()
```

```python
_, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(c_2016)
ax[1].imshow(c_2017)
```

```python
c_hist_2016 = np.array([image_hist(c_2016, i) for i in range(3)]).reshape(3, 256)
c_hist_2017 = np.array([image_hist(c_2017, i) for i in range(3)]).reshape(3, 256)
c_hist_2016.shape, c_hist_2017.shape
```

```python
_, ax = plt.subplots(3, 1, figsize=(10,10))

for i,c in enumerate(["blue", "green", "red"]):
    ax[i].set_title("{} 2016 vs. {} 2017".format(c,c))
    ax[i].plot(c_hist_2016[i, :], label="2016")
    ax[i].plot(c_hist_2017[i, :], color="r", label="2017")
    ax[i].legend()
```
